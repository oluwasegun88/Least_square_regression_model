```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
hlt = pd.read_csv("health_data.csv")
hlt.head()
```

```python
hlt.isna().sum()
```

```python
hlt_records_dropped = hlt.dropna(axis=0, how='any')
hlt_records_dropped.info()
```

```python
hlt_records_dropped.isna().sum()
```

```python
hlt_records_dropped.info()
```

```python
hlt_records_dropped.describe()
```

```python
hlt_records_dropped = hlt_records_dropped.drop_duplicates()
hlt_records_dropped
```

```python
df1 = hlt_records_dropped.drop(columns=['binnedinc', 'geography'], axis=1)
df1.info()
```

```python
y =df1 ["target_deathrate"]
x = df1.drop(columns=["target_deathrate"])
```

```python
x
```

```python
y
```

```python
df1['target_deathrate'].value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
print('shape of training set: ',X_train.shape)
print('shape of test set: ',X_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
pred=model.predict(X_test)
```

```python
predictions = pred.reshape(-1,1)
```

```python
print('MSE : ', mean_squared_error(y_test,predictions))
print('RMSE : ', np.sqrt(mean_squared_error(y_test,predictions)))
```
